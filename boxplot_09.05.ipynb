```python
#Importar bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Carregar a planilha .csv que deesja
df1 = pd.read_csv('produtividades.csv')
```

```python
dados_por_talhao = [df1[coluna].dropna().tolist() for coluna in df1.columns]

# Criando uma figura e eixos do Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))

# Criando o boxplot
ax.boxplot(dados_por_talhao)

# Definindo rótulos e título
ax.set_title('Boxplot da Produtividade por Talhão')
ax.set_xlabel('Talhões')
ax.set_ylabel('Produtividade (t ha-¹)')
ax.set_xticklabels(df1.columns, rotation=45)


# Mostrando o gráfico
plt.show()
```
